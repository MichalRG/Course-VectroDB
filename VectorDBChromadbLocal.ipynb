```python
import chromadb
```

```python
import pandas as pd
```

```python
df = pd.read_csv("medium_post_titles.csv")
df
```

```python
df = df.dropna()
df
```

```python
df = df[-df["subtitle_truncated_flag"]]
df
```

```python
# topics_of_interest = ["data-science"]
topics_of_interest = ["artificial-intelligence", "machine-learning", "data-science"]
```

```python
df=df[df["category"].isin(topics_of_interest)]
df
```

```python
df["text"]=df["title"]+df["subtitle"]
df
```

```python
df["meta"] =df.apply( lambda x: {
    "text": x["text"],
    "category": x["category"]
}, axis=1)
df
```

```python
df.head(2)
```

```python
### CHROMA DB SETUP
```

```python
chroma_client = chromadb.Client() #no argument = in memeory db
```

```python
#collection creation
article_collection = chroma_client.create_collection(name="medium-article")
```

```python
df.index.tolist()
```

```python
#inserting data
article_collection.add(
    ids=[f"{x}" for x in df.index.tolist()], # vector ids - strings
    documents=df["text"].tolist(), #each row/text in database is document
    metadatas = df["meta"].tolist()
)
```

```python
#article_collection.delete() #removing collection
#article_collection.upsert() #takie samo jak add ale tutaj dodaje i update'uje jezeli jest zmiana
```

```python
#vector query
qry_str =  "best data science library?"
```

```python
article_collection.query(query_texts=qry_str, n_results=1) #n_result means how many results we get
```

```python
qry_str2 = "best data ai library?"
```

```python
article_collection.query(query_texts=qry_str2, n_results=2)
```

```python
# tunning chroma db with custom setting
from chromadb.config import Settings
```

```python
chroma_client2 = chromadb.Client(Settings(
    persist_directory="medium-chroma-db", #place where db will be saved in current working directory
    chroma_db_impl="duckdb-parquet" #default value
))
article_collection.delete()
article_collection = chroma_client2.create_collection(name="medium_article")
```
